```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets, linear_model

house_price = [245, 312, 279, 308, 199, 219, 405, 324, 319, 255 ]
size = [1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700]

size2=np.array(size).reshape(-1,1)
house_price2=np.array(house_price).reshape(-1,1)
print(size2)
print(house_price2)

regr = linear_model.LinearRegression()
regr.fit(size2, house_price2)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)

def graph(regr, size, house_price):
    plt.scatter(size, house_price, color='blue')
    plt.plot(size, regr.predict(size2), color='red', linewidth=4)

    plt.xticks(())
    plt.yticks(())
    plt.show()
graph(regr, size, house_price)
graph(regr, size2, house_price2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('ggplot')
from sklearn.cluster import KMeans

x = [1, 5, 1.5, 8, 1, 9]
y = [2, 8, 1.8, 8, 0.6, 11]

# plt.scatter(x, y)
#plt.show()

X = np.column_stack((x, y)).astype('int64')
print(X)
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)

centroids = kmeans.cluster_centers_
labels = kmeans.labels_
print(centroids)
print(labels)

colors = ["g.","r."]
for i in range(len(X)):
  print("coordinate:",X[i], "label:", labels[i])
  plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize = 10)

plt.scatter(centroids[:, 0],centroids[:, 1], marker = "x", s=150, linewidths = 5, zorder = 10)
plt.show()
```
